```python
import os
import json

def load_json_files_from_folder(folder_path):
    json_files = [f for f in os.listdir(folder_path) if f.endswith('.json')]
    data_list = []

    for json_file in json_files:
        file_path = os.path.join(folder_path, json_file)
        with open(file_path, 'r') as f:
            data = json.load(f)
            data_list.append(data)

    return data_list

folder_path = '.'
all_json_data = load_json_files_from_folder(folder_path)

# # Print or process the loaded JSON data
# for data in all_json_data:
#     print(data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_score(score):
    
    fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
    
    for i in range(3):
        ax = axes[i]
        for data in all_json_data:  
            ax.bar(data['name'], data['results'][i]['scores'][score])
            print(data['name'], data['results'][i]['scores'][score])
            ax.set_title(data['results'][i]['targetLabel'])
            ax.tick_params(axis='x', rotation=90)
        print()
    fig.suptitle(score)
    plt.tight_layout()
    plt.show()

plot_score('accuracy')
#plot_score('f1')
#plot_score('roc_auc')
#plot_score('pr_auc')
#plot_score('mcc')
#plot_score('cohen_kappa')
```
